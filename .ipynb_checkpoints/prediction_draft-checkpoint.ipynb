```python
import pandas as pd
from pandas.tseries.offsets import MonthBegin
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = "https://www.cba.am/stat/stat_data_eng/8_policy_rates_eng.xlsx"

# Read the Excel file directly using read_excel 
refinancing_rate = pd.read_excel(url)

# Display the data
print(refinancing_rate)
```

```python
refinancing_rate = refinancing_rate.iloc[2:]
refinancing_rate.columns = ["Time_Period", "Refinancing_Rates", "Deposits_Attracted", "Lombard_Repo1"]

refinancing_rate = refinancing_rate[['Time_Period', 'Refinancing_Rates']]

refinancing_rate.dropna(inplace=True)

# Assuming refinancing_rate is your DataFrame and it's already loaded

# First, strip leading/trailing whitespace from the split date strings
refinancing_rate['Start_Date_Str'] = refinancing_rate['Time_Period'].str.split('-').str[0].str.strip()

# Then, convert these cleaned strings to datetime format
refinancing_rate['Start_Date'] = pd.to_datetime(refinancing_rate['Start_Date_Str'], format='%d.%m.%Y', errors='coerce')

# Optionally, you can drop the 'Start_Date_Str' column if you don't need it anymore
refinancing_rate.drop('Start_Date_Str', axis=1, inplace=True)

# You can also drop the original 'Time_Period' column if it's no longer needed
refinancing_rate.drop('Time_Period', axis=1, inplace=True)


# Set 'Start_Date' as the index
refinancing_rate.set_index('Start_Date', inplace=True)

# Resample to monthly frequency and forward fill
monthly_df = refinancing_rate.resample('M').ffill()
```

```python
print(refinancing_rate)
```

```python
refinancing_rate.index = refinancing_rate.index.to_period('M')
print(refinancing_rate)
```

```python
from dateutil.relativedelta import relativedelta

# Since refinancing_rate.index is a PeriodIndex, convert it to DateTimeIndex
refinancing_rate.index = refinancing_rate.index.to_timestamp()

# Assuming the DataFrame refinancing_rate is now correctly indexed by DateTime

# Calculate the start and end dates
start_date = refinancing_rate.index.min()
end_date = refinancing_rate.index.max()

# Initialize a new DataFrame to hold the result, if necessary
# Note: Based on your current process, you might not need to reinitialize refinancing_rate

# Generate all months between start and end
current_date = start_date

while current_date <= end_date:
    if current_date not in refinancing_rate.index:
        # Insert a new row with NaN for Refinancing_Rates at the current_date
        # This method avoids directly concatenating within the loop, which can be inefficient
        refinancing_rate.at[current_date, 'Refinancing_Rates'] = pd.NA
    # No else block needed; if current_date is in the index, it's already part of refinancing_rate

    # Move to the next month using relativedelta to ensure compatibility
    current_date += relativedelta(months=1)

# After updating, it's a good practice to sort the DataFrame by the index
refinancing_rate.sort_index(inplace=True)

# Display the updated DataFrame
print(refinancing_rate.head(15))  # Adjust as needed
print(refinancing_rate.tail())
```

```python
# Fill NaN values with the previous non-NaN value in the Refinancing_Rates column
refinancing_rate['Refinancing_Rates'] = refinancing_rate['Refinancing_Rates'].fillna(method='ffill')

# Display the DataFrame to verify the NaN values are filled
print(refinancing_rate.head(15))  # Adjust as needed
print(refinancing_rate.tail())
```

```python
refinancing_rate.index = refinancing_rate.index.to_period('M')
print(refinancing_rate)
```

```python
# Read the Excel file directly using read_excel 
dolarization = pd.read_excel("https://www.cba.am/stat/stat_data_eng/16_Dollarization_overall_eng.xlsx")

# Display the data
print(dolarization)
```

```python
# Assuming 'df' is your DataFrame
# Select only the third column, keeping it as a DataFrame
dolarization = dolarization.iloc[:, [0, 2]]
dolarization = dolarization.iloc[2:]


dolarization.dropna(inplace=True)
dolarization.rename(columns={'Level of dollarization in RA*': 'Time_Period', 'Unnamed: 2': 'Dolarization/D2'}, inplace=True)


# Display the third column DataFrame
print(dolarization)
```

```python
# Convert 'Time_Period' to datetime format if it's not already
dolarization['Time_Period'] = pd.to_datetime(dolarization['Time_Period'])

# Format 'Time_Period' to "YYYY-MM" string format
dolarization['Time_Period'] = dolarization['Time_Period'].dt.strftime('%Y-%m')

# Display the DataFrame to verify the changes
print(dolarization)
```

```python
# Example: If the total decrease from 2012 to 2024 is about 20 units over 12 years,
# and you want a slower decrease rate for the years 2003 to 2012,
# you might assume a decrease of 0.5 units per year for those earlier years.

# Calculate the starting value for December 2003
# Assume the value in January 2012 is the earliest known value
earliest_known_value = dolarization['Dolarization/D2'].iloc[0]  # Assuming dolarization is sorted by Time_Period

# Calculate how many years back we're going from 2012 to 2003 (9 years)
years_back = 2012 - 2003

# Assuming a decrease of 0.5 per year for those years
decrease_rate = 0.5
estimated_start_value_2003 = earliest_known_value + (years_back * decrease_rate)

# Now, create the missing DataFrame for 2003-12 to 2011-12 with these adjusted values
missing_years = pd.date_range(start='2003-12-01', end='2011-12-01', freq='MS')
missing_values = [estimated_start_value_2003 - (decrease_rate * i / 12) for i in range(len(missing_years))]

missing_df = pd.DataFrame({'Time_Period': missing_years, 'Dolarization/D2': missing_values})

# Concatenate this with your original data and sort
filled_dolarization = pd.concat([missing_df, dolarization], ignore_index=True)

# Convert 'Time_Period' column to datetime
filled_dolarization['Time_Period'] = pd.to_datetime(filled_dolarization['Time_Period'], errors='coerce')

filled_dolarization.sort_values(by='Time_Period', inplace=True)

# Ensure Time_Period is the correct type if needed
filled_dolarization['Time_Period'] = pd.to_datetime(filled_dolarization['Time_Period'])

# Format 'Time_Period' to "YYYY-MM" string format
filled_dolarization['Time_Period'] = filled_dolarization['Time_Period'].dt.strftime('%Y-%m')

# Display the DataFrame to verify the changes
print(filled_dolarization)
```

```python
print("filled_dolarization index name:", filled_dolarization.index.name)
print("refinancing_rate index name:", refinancing_rate.index.name)

# If refinancing_rate index name is 'Start_Date' or similar, align it with filled_dolarization
if refinancing_rate.index.name != 'Time_Period':
    refinancing_rate.index.names = ['Time_Period']  # This aligns the index name without changing the underlying data

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
data['Time_Period'] = pd.to_datetime(data['Time_Period'])
data['Year'] = data['Time_Period'].dt.year
data['Month'] = data['Time_Period'].dt.month
```

```python
# Prepare features and target variable
X = data[['Year', 'Month','Dolarization/D2']]  # Add more features if needed
y = data['Refinancing_Rates']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on the testing set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')
```

```python
# Function to handle mixed types in 'Time_Period' and standardize to datetime format
def standardize_date(date_entry):
    if isinstance(date_entry, str):  # Check if the entry is a string
        # Attempt to split and format based on expected non-standard formats
        try:
            # Split if there's a space and take the first part, then convert to datetime
            date_str = date_entry.split(' ')[0]
            if '-' in date_str and len(date_str) <= 7:  # Expected to catch 'Dec-11' like formats
                return pd.to_datetime(date_str, format='%b-%y')
            else:
                return pd.to_datetime(date_str)  # Handle other string representations
        except ValueError:
            # Log or handle dates that couldn't be parsed
            return pd.NaT
    elif isinstance(date_entry, pd.Timestamp):
        return date_entry  # If already a Timestamp, no conversion needed
    else:
        return date_entry  # For any other types not accounted for

# Apply the function to the 'Time_Period' column to standardize all entries to datetime
international_reserves['Time_Period'] = international_reserves['Time_Period'].apply(standardize_date)

# Print the updated DataFrame to verify changes
print(international_reserves)
```
